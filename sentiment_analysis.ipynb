# Function for data cleaning

```python
import json
import string
def clean_text(text):
    #making a function to convert the reviews into lower case,removing their punctuation and remove stop words
    clean_text = text.lower().translate(str.maketrans('', '', string.punctuation))
    clean_text = ' '.join([word for word in clean_text.split() if word not in stop_words])
    return clean_text
```

# Defining stop words

# Cleaning and loading the data set

```python
#making two lists
#one for storing original reviews being used later and the other for storing cleaned reviews without stop words and punctuation
cleaned_data_list = [] 
original_data_list = []  
#opening json file, making json object and loading it. 
#Making different dictionaries of multiple objects in json file and appending these dictionaries to the list
with open('Cell_Phones_and_Accessories_5.json') as f:
    for jsonObj in f:
        data_dict = json.loads(jsonObj)
        if 'reviewText' in data_dict and 'overall' in data_dict:
            original_data_list.append(data_dict.copy())
            data_dict['reviewText'] = clean_text(data_dict['reviewText'])
            del data_dict["reviewerID"]
            del data_dict["asin"]
            del data_dict["helpful"]
            del data_dict["unixReviewTime"]
            del data_dict["reviewTime"]
            del data_dict["summary"]
            cleaned_data_list.append(data_dict)
```

# Defining positive and negative words with their scores in a dictionary 

```python
#making two lists
#one to store ngative reviews and one to store positive reviews
positive_reviews=[]
negative_reviews=[]
#iterating through the previous list, storing ratings in a variable
#putting reviews with ratigns above or equal to 3 in one list and negative reviews in the other list
for data in cleaned_data_list:
    product_ratings = data['overall']
    reviews = data['reviewText']
    if product_ratings >= 3:  
        positive_reviews.append(reviews)
    elif product_ratings <=2: 
        negative_reviews.append(reviews)
#display(positive_reviews)
#display(negative_reviews)
```

# extracting most common words from both negative and positive reviews

```python
#making a function for counting trigrams in the cleaned review text
def counting_bigrams(review_text):
    #storing the frequnecy of each bigram in a dictionary
    frequency_of_bigrams = {}
    reviews = review_text.split('\n') 
    #iterating through each review
    for review in reviews:
        #converting the reviews into lowercase and splitting them and storing them in a list
        words = review.lower().split()
        #iterating throught the length of list
        for i in range(len(words) - 1):
            # Create a trigram by joining current word and the next word
            bigram = ' '.join(words[i:i+3])
            # Update the frequency count for the trigram
            frequency_of_bigrams[bigram] = frequency_of_bigrams.get(bigram, 0) + 1
    #return dictionary which has the frequency of trigrams
    return frequency_of_bigrams
#create a list to store common words that occur in positive comments
positive_reviews_commonwords = []

#counting trigrams in all the positive reviews
positive_reviews_bigrams = counting_bigrams('\n'.join(positive_reviews))

#extracting common trigrams across the positive reviews
for bigram, frequency in positive_reviews_bigrams.items():
    if frequency > 20:
        positive_reviews_commonwords.append(bigram)

print(positive_reviews_commonwords)
```

```python
negative_reviews_commonwords = []

# Count trigrams across all negative reviews
negative_reviews_bigrams = counting_bigrams('\n'.join(negative_reviews))

# Filter out common trigrams in all the negtaive reviews
for bigram, frequency in negative_reviews_bigrams.items():
    if frequency > 10:
        negative_reviews_commonwords.append(bigram)

print(negative_reviews_commonwords)
```

```python
import re
#making a function for entiment analysis
def rule_based_sentiment_analysis(text):
    #Converting the reivew text to lowercase
    text_lower = text.lower()

    #Count occurrences of positive and negative keywords and storing them in variables
    positive_count = sum([len(re.findall(keyword, text_lower)) for keyword in positive_words])
    negative_count = sum([len(re.findall(keyword, text_lower)) for keyword in negative_words])

    # Determine sentiment based on keyword counts
    if positive_count > negative_count:
        #if positive words in a comment are occuring more then negative words in a comment, then return positvie
        return "Positive"
    
    elif positive_count < negative_count:
        #if negative words are occuring in a comment more then positive words, then return negative
        return "Negative"
    else:
        #if both are eually occuring, return neutral
        return "Neutral"
#for data in original_data_list:
    #for review in data['reviewText'].split('\n'):
        #sentiment = rule_based_sentiment_analysis(review)
        #display(review + ': ' + sentiment)
```

```python
with open("sentiment_analysis_results.txt", "w") as file:
    # Loop through original data list and analyze each review
    for data in original_data_list:
        for review in data['reviewText'].split('\n'):
            # Perform sentiment analysis
            sentiment = rule_based_sentiment_analysis(review)
            # Write the review and sentiment to the file
            file.write(review + ': ' + sentiment + '\n')
```
